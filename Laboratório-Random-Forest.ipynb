```python
import pandas as pd
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Como você mapeou o volume no Docker, os arquivos estão acessíveis na pasta raiz
print("Carregando datasets...")
df_treino = pd.read_csv('./dataset-treated/cicids2017_tratado_normalizado.csv')
df_teste = pd.read_csv('./dataset-treated/cicids2017_tratado_normalizado_teste.csv')

# Separando Features (X) e Labels (y)
X_train = df_treino.drop('Label', axis=1)
y_train = df_treino['Label']

X_test = df_teste.drop('Label', axis=1)
y_test = df_teste['Label']

print("Dados carregados com sucesso!")
```

```python
# Instancia o modelo Random Forest
# n_estimators=100 significa que ele criará 100 árvores de decisão
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

print("Iniciando o treinamento do Random Forest...")
inicio_treino = time.time()

# O comando .fit() é o mecanismo onde o aprendizado (criação das árvores) acontece
rf_model.fit(X_train, y_train)

fim_treino = time.time()
tempo_treinamento = fim_treino - inicio_treino

print(f"Treinamento concluído! Tempo total: {tempo_treinamento:.2f} segundos.")
```

```python
print("Iniciando a classificação dos dados de teste...")
inicio_teste = time.time()

# O modelo agora usa as regras aprendidas para prever os dados que ele nunca viu
previsoes = rf_model.predict(X_test)

fim_teste = time.time()
print(f"Classificação concluída! Tempo de previsão: {fim_teste - inicio_teste:.2f} segundos.\n")

# Gerando as métricas exigidas pela metodologia do TCC
print("=== MATRIZ DE CONFUSÃO ===")
print(confusion_matrix(y_test, previsoes))
print("\n=== RELATÓRIO DE MÉTRICAS (Precisão, Recall, F1-Score) ===")
print(classification_report(y_test, previsoes))
```
